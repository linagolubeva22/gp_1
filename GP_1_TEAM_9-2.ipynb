Первая оценка — это оценка для всего проекта в целом. Она ставится комиссией, исходя из глубины проработки командой предметной области, сложности применяемых в проекте инструментов, а также прочих аспектов, связанных с содержанием выполняемого в проекте ТЗ. Вторая оценка — это индивидуальная оценка каждого участника команды. Именно эта оценка выставляется впоследствии в ведомость. Индивидуальная оценка участника команды определяется комиссией, исходя из того, насколько участник ориентируется в выполненном командой проекте, как он отвечает на адресованные персонально ему вопросы комиссии, насколько в принципе прослеживается его роль в проекте, а также исходя из других аспектов, связанных с участием конкретного человека в деятельности команды. Индивидуальная оценка не может превосходить оценку команды за проект в целом (первую оценку), однако, в случае когда участник спокойно и корректно отвечает на все адресованные ему вопросы, свободно ориентируется в материале и содержании выполненного проекта, а также когда его роль в работе команды не вызывает сомнений, — в таком случае индивидуальная оценка участника становится равной общей оценке команды за проект (т.е. в идеальном случае — вторая оценка равна первой оценке).

# Первичный обзор и описание переменных сырого датасета

**Общая характеристика датасета**

На первом этапе исследования был проведён первичный осмотр исходного набора данных

1. **title**

Строковая переменная, содержащая название тайтла.

На данном этапе сложно сделать однозначный вывод о происхождении дубликатов.
С одной стороны, часть повторов может быть связана с техническими особенностями выгрузки данных.
С другой стороны, нельзя исключать, что они отражают специфику формирования выборки, например попадание одного и того же тайтла в несколько рекомендационных цепочек или категорий

Хм, а есть ли еще другие дубли? А вдруг есть дубли с одинаковым названием и годом, но разным рейтингом?


```python
df.duplicated(subset=["title", "release_year"]).sum()
```

```python
dups = df[df.duplicated(subset=['title', 'release_year'], keep=False)]
different_rating = (dups.groupby(['title', 'release_year'])['rating'].nunique())
different_rating[different_rating > 1]
```

```python
problem_titles = different_rating[different_rating > 1].index.get_level_values('title')
df[df['title'].isin(problem_titles)].sort_values('title')
```

Ага, попались, значит гипотеза подтвердилась

Но!

Значит что при этом это может указывать на возможные различия в метаданных или способе агрегирования информации

```python
dups = df[df.duplicated(subset=['title', 'release_year'], keep=False)]
dups.groupby('rating').size().sort_values(ascending=False)
```

```python
before = df.groupby('rating').size().rename('до')
after = (df.drop_duplicates(subset=['title', 'release_year']).groupby('rating').size().rename('после'))

result = pd.concat([before, after], axis=1)
result['удалено'] = result['до'] - result['после']
result['доля %']  = (result['удалено'] / result['до'] * 100).round(1)
result.sort_values('удалено', ascending=False)
```

В датасете обнаружено 501 дублирующихся строк из 1000 (50,1%). Все дубликаты полностью идентичны — различий в оценках нет ни у одного шоу (кроме того самого 1 случая). Больше всего дублей по абсолютному числу в группах TV-14 (128), PG (94) и G (85) — просто потому что эти группы самые многочисленные. По доле лидирует G (61.6%) и PG (55.3%). Меньше всего пострадали малочисленные группы PG-13 (20%) и R (26.3%)

Тем не менее полученный результат позволяет предположить, что временной контекст может частично объяснять вариацию пользовательских оценок.
В связи с этим дальнейшее заполнение пропусков целесообразно проводить с учетом временных когорт

На первом этапе пропущенные значения будут восстановлены медианами внутри более узких когорт, сформированных по сочетанию года выпуска и рейтинговой категории.

Удалили. Теперь давайте создадим признак, в который проставим тип рейтинга: TV или not TV, то есть это рейтинг тв шоу или не тв шоу

Наши рейтинговые группы намекают на тип контента: все, что начинается с TV, относится к тв рейтингу, остальное к не тв шоу. Воспользуемся этой информацией

```python
movie_rating = ['G', 'PG', 'PG-13', 'R'] # заполняем сначала все как тв, а потом по локу обращаемся к конкреным ячейкам
df['rating_type'] = 'TV'
df.loc[df['rating'].isin(movie_rating), 'rating_type'] = 'not TV'
```

```python
df.head(3)
```

Вообще стоит дополнительно выделить признак по возрастной категории, так как рейтинговые группы достаточно сложно воспринимать

В нашем датасете у нас можно выделить несколько возрастных категорий:

1. Babies(0+) - подходит для любого возраста. Абсолютно безопасный контент. Сюда отнесем G, TV-G, TV-Y
2. Kids(7-13) - подходит с 7 до 13 лет. Чуть более сложный контент, в котором может быть "плохой" юмор и небольшое насилие. Сюда включаем PG, TV-PG, TV-Y7, TV-Y7-FV
3. Teenagers(14-17) - уже сильно более взрослый контент, с взрослыми темами и насилием. Сюда отнесем следующие рейтинговые категории: PG-13, TV-14, R
4. Adults(18+) - Взрослый контент для совершеннолетних. TV-MA

```python
df['age_group'] = 'no age group'

babies_group = ['G', 'TV-G', 'TV-Y']
kids_group = ['PG', 'TV-PG', 'TV-Y7', 'TV-Y7-FV']
teenagers_group = ['PG-13', 'TV-14']
adults_group = ['TV-MA', 'R']

# заполняем аналогично предыдущей ячейке, если значение в массиве, то ставим соответсвующую возрастную категорию
df.loc[df['rating'].isin(babies_group), 'age_group'] = 'Babies(0+)'
df.loc[df['rating'].isin(kids_group), 'age_group'] = 'Kids(7-13)'
df.loc[df['rating'].isin(teenagers_group), 'age_group'] = 'Teenagers(14-17)'
df.loc[df['rating'].isin(adults_group), 'age_group'] = 'Adults(18+)'
```

```python
df.sample(5)
```

Отлично, feature engineering позанимались, теперь мы можем понимать на какой возраст ограничения. Теперь интересно посмотреть, какая средняя оценка рекомендации в каждой из возрастных групп. Давайте для этого сделаем сводную таблицу и изучим что в ней происходит

```python
# делаем сводную таблицу, считаем среднее значение по user_rating_size только по возрасным категориям(index) + сорируем
rating_groups_score = pd.pivot_table(df, values = 'user_rating_score',
                                     index = 'age_group', aggfunc = 'mean').sort_values(by = 'user_rating_score')
rating_groups_score
```

```python
import matplotlib.pyplot as plt
```

```python
categories = ['Babies(0+)', 'Kids(7-13)', 'Teenagers(14-17)', 'Adults(18+)']
rating_groups_score = rating_groups_score.reindex(categories) # делаем, чтобы категории шли в нужном порядке
colors = ['#7B001C', '#9B111E', '#C41E3A', '#FF2B2B']
plt.figure(figsize = (14, 6))
plt.barh(rating_groups_score.index, rating_groups_score['user_rating_score'], color = colors) # горизонтальная барчар

# подписываем столбцы справа
plt.text(79, 0, round(rating_groups_score.loc['Babies(0+)','user_rating_score'], 2), fontweight = 'bold')
plt.text(83, 1, round(rating_groups_score.loc['Kids(7-13)','user_rating_score'], 2), fontweight = 'bold')
plt.text(83, 2, round(rating_groups_score.loc['Teenagers(14-17)','user_rating_score'], 2), fontweight = 'bold')
plt.text(85, 3, round(rating_groups_score.loc['Adults(18+)','user_rating_score'], 2), fontweight = 'bold')

plt.title('Средняя оценка контента в зависимости от возрастной категории', fontweight = 'bold')
plt.ylabel('Возрастная категория')
plt.xlabel('Средняя оценка')
```

Отсюда мы видим, что

1) контент для самых маленьких имеет самую низкую оценку подходящести пользователю

2) Контент для детей и подростков в среднем подходит одинаково пользователю

3) Больше всего в среднем пользователю подходит контент с возрастным ограничением 18+

В целом, разница между сегментами небольшая, но явно можем сказать, что более взрослый контент подходит пользователю больше, чем детский контент

## Влияние жанра на пользовательские оценки контента

Для начала посмотрим на сам столбец genres, как у нас выглядят данные и есть ли пропуски




```python
genres_df = df_1.copy()
genres_df['genres'].isna().sum()
```

```python
genres_df['genres'].head(20)
```

Пропусков у нас нет - это просто супер!

А так же мы видим, что значения в столбце представлены в виде списка жанров и для каждого фильма/сериала их может быть несколько

В дальнейшем анализе мы будем рассматривать каждый жанр отдельно, а не анализировать их комбинации. Это связано с тем, что комбинации жанров образуют очень большое количество уникальных сочетаний, что сильно усложняет анализ и делает результаты менее наглядными

Кроме того, один и тот же жанр может встречаться в разных сочетаниях.

И если анализировать такие сочетания как отдельные категории, то влияние самого жанра будет сложно оценить.

Поэтому далее для удобства анализа преобразуем столбец жанров так, чтобы каждый рассматривался как отдельное наблюдение

```python
genres_df['genres'] = genres_df['genres'].str.strip('[]').str.replace("'", "") #убираем квадратные скобки из списка
genres_df['genres'] = genres_df['genres'].str.split(', ') #удаляем кавычки вокруг названий жанров
genres_df = genres_df.explode('genres') #каждый жанр становится отдельной строкой,а остальные данные о фильме дублируются
genres_df['genres'].head(10)
```

После подготовки данных можно перейти к анализу жанров

Посмотрим, какие жанры представлены в датасете, сколько получилось уникальных значений и как часто встречается каждый из них

```python
genres_df['genres'].nunique()
```

```python
genres_df['genres'].value_counts()
```

По результатам видно, что количество уникальных жанров в выборке относительно небольшое, поэтому данный признак удобно анализировать дальше

Поскольку в датасете представлены и фильмы, и сериалы, далее целесообразно рассматривать распределение жанров отдельно по типам контента. Это позволит увидеть, отличаются ли жанровые предпочтения у фильмов и у сериалов

```python
genres_cnt = (genres_df.groupby(['genres', 'true_type']).size().unstack())
```

```python
colors = ['#9B111E', '#FF2B2B']

genres_cnt.plot(kind='bar',color=colors, figsize=(16,7))

plt.title('Распределение жанров по типу контента', fontweight='bold')
plt.xlabel('Жанр')
plt.ylabel('Количество')

plt.legend(title='Тип контента',labels=['Фильмы', 'ТВ-шоу'])

plt.show()
```

Так, среди фильмов наиболее распространенными у пользователя являются жанры `комедия, семейный, приключения и мультфильм`, также заметную долю занимают `драма и криминал`, которые традиционно популярны в полнометражном формате

В то же время для сериалов характерна более высокое количество жанров `драма, детектив, триллер и криминал`

После анализа распределения жанров можно перейти к изучению их связи с целевой переменной `user_rating_score`

Если предыдущий этап показал, какие жанры чаще встречаются в датасете, то теперь важно понять, различаются ли пользовательские оценки в зависимости от жанра

Перед построением графика исключим жанры, которые встречаются реже 5раз, для того, чтобы результаты были более интерпретируемыми.

## Анализ влияния выручки на пользовательскую оценку

В ходе предварительного изучения данных было выявлено, что информация о выручке представлена преимущественно для фильмов, кассовые сборы для которых являются стандартным и широко публикуемым показателем, а коммерческий успех сериалов обычно оценивается по другим метрикам

Далее рассмотрим столбец с выручкой `fees`, на этом этапе нам важно понять есть ли в нем пропуски и какие значения в целом имеются






Сначала проверим количество пропущенных значений

```python
df_1[df_1['true_type'] == 'Movie']['fees'].isna().sum()
```

```python
df_1[df_1['true_type'] == 'Movie']['fees'].isna().mean()* 100
```

В столбце выручки обнаружено около 42% пропусков. Поскольку целью анализа является изучение влияния кассовых сборов на пользовательскую оценку, наблюдения без информации о выручке не могут быть использованы. В связи с этим они были исключены из дальнейшего анализа

```python
fees_df = df_1[['fees', 'user_rating_score']].dropna()
```

```python
fees_df['fees'].isna().sum()
```

После анализа пропущенных значений рассмотрим, как в целом распределена выручка в датасете.

Для этого построим боксплот, который позволит оценить медиану, разброс значений и наличие выбросов

```python
plt.figure(figsize=(10,5))
plt.boxplot(fees_df['fees'])
plt.title('Boxplot выручки')
plt.xlabel('Выручка')
plt.show()
fees_df['fees'].describe()
```

Интересно, боксплот показывает наличие значительного количества выбросов в правой части распределения и это значит, что некоторое число фильмов имеет крайне высокие кассовые сборы по сравнению с основной массой

Однако такие значения отражают реальную структуру рынка киноиндустрии и поэтому их удаление выбросов может привести к потере важной информации, поэтому дальше сгруппируем выручку по категориям (в нашем случае по квартилям)

После формирования категорий выручки можно посмотреть, как меняется целевая переменная в зависимости от уровня кассовых сборов. Для этого рассчитаем среднее значение оценки в каждой категории, а затем визуализируем результат на графике

```python
fees_mean = fees_df.groupby('fees_category')['user_rating_score'].mean()
fees_mean
```

```python
plt.figure(figsize=(12,6))

plt.bar(fees_mean.index,fees_mean.values,color='#7B001C')

plt.title('Средняя пользовательская оценка в зависимости от категории выручки',fontweight='bold')

plt.xlabel('Категория выручки')
plt.ylabel('Средний рейтинг')

plt.show()
```

**ВЫВОДЫ**

По построенному графику можно заметить, что тайтлы с высокой выручкой в среднем получают более высокие оценки пользователей по сравнению с фильмами с низкими кассовыми сборами

Наименьшие значения средней оценки наблюдаются у фильмов с низкой и ниже средней выручкой, тогда как у фильмов, относящихся к категории выше средней и высокой выручки, оценки заметно выше

При этом различия между соседними категориями выручки не являются уж слишком значительными, что говорит о том, что данный фактор должен рассматриваться в совокупности с другими характеристиками контента

По графику видно, что в ранние годы каталог почти полностью состоял из фильмов, однако со временем доля сериалов постепенно росла: сначала структура стала более смешанной, а в последние годы TV Show уже занимают сопоставимую или даже большую долю, чем Movie. В целом это показывает сдвиг каталога от преимущественно фильмов к более сбалансированной, а затем и более сериальной структуре, при этом резкие колебания в отдельных ранних годах стоит трактовать осторожно из-за малого числа наблюдений.

```python

plot_df = df[["countries", "true_type"]].dropna().copy()

agg = (plot_df.groupby(["countries", "true_type"]).size().reset_index(name="count"))

wide = (agg.pivot(index="countries", columns="true_type", values="count").fillna(0).reset_index())

wide["Total"] = wide["Movie"] + wide["TV Show"]

iso_map = {
    "Австралия": "AUS",
    "Бразилия": "BRA",
    "Великобритания": "GBR",
    "Германия": "DEU",
    "Дания": "DNK",
    "Индия": "IND",
    "Ирландия": "IRL",
    "Испания": "ESP",
    "Канада": "CAN",
    "Корея Южная": "KOR",
    "Малайзия": "MYS",
    "Мексика": "MEX",
    "США": "USA",
    "Франция": "FRA",
    "Япония": "JPN",
}

wide["iso"] = wide["countries"].map(iso_map)


wide["Movie_log"] = np.log1p(wide["Movie"])
wide["TV_Show_log"] = np.log1p(wide["TV Show"])
wide["Total_log"] = np.log1p(wide["Total"])

zmax = max(wide["Movie_log"].max(),wide["TV_Show_log"].max(),wide["Total_log"].max())

tick_vals = np.log1p([1, 3, 10, 30, 100])
tick_text = ["1", "3", "10", "30", "100"]

fig = go.Figure()
fig.add_trace(go.Choropleth(locations=wide["iso"],
                            z=wide["Movie_log"],
                            text=wide["countries"],
                            locationmode="ISO-3",
                            colorscale="Reds",
                            zmin=0,
                            zmax=zmax,
                            colorbar=dict(
                                title="Количество",
                                tickvals=tick_vals,
                                ticktext=tick_text
                            ),
                            customdata=wide[["Movie", "TV Show", "Total"]],
                            hovertemplate=(
                                "<b>%{text}</b><br>"
                                "Фильмы: %{customdata[0]}<br>"
                                "Сериалы: %{customdata[1]}<br>"
                                "Всего: %{customdata[2]}"
                            )
                        )
                    )
buttons = [dict(label="Фильмы",method="update",
        args=[
            {"z": [wide["Movie_log"]],
             "colorscale": ["Reds"],
             "zmin": [0],
             "zmax": [zmax]},
            {"title": "Карта мира: количество фильмов по странам"}
        ]),

    dict(label="Сериалы",method="update",
        args=[{"z": [wide["TV_Show_log"]],
               "colorscale": ["Blues"],
               "zmin": [0],
               "zmax": [zmax]},
            {"title": "Карта мира: количество сериалов по странам"}
        ]),

    dict(label="Всего", method="update",
        args=[{"z": [wide["Total_log"]],
                "colorscale": ["Purples"],
                "zmin": [0],
                "zmax": [zmax]},
            {"title": "Карта мира: общее количество тайтлов по странам"
        }])]



fig.update_layout(title="Карта мира: количество фильмов по странам",
            updatemenus=[dict(type="dropdown",
                        direction="down",
                        buttons=buttons)])

fig.show()
```

По карте видно, что основная часть контента сосредоточена в США, которые заметно лидируют по количеству и фильмов, и тайтлов в целом. При этом производство распределено неравномерно: помимо США, более заметный вклад дают Канада, Великобритания, Австралия, Франция, Индия, Япония и Мексика, тогда как у большинства остальных стран объёмы существенно ниже. Логарифмическая шкала помогает увидеть не только лидера, но и различия между странами со средними и малыми значениями, поэтому карта показывает, что контент представлен в нескольких регионах мира, однако ядро каталога явно сконцентрировано в ограниченном числе стран.
